```python
#imports
import cv2
import numpy as np
import matplotlib.pyplot as plt

print("OpenCV version:", cv2.__version__)
```

```python
#load


img = cv2.imread("car.jpg")

if img is None:
    print("Upload an image named sample.jpg")
else:
    print("Image loaded successfully")


```

```python
#image and properties


plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")


print("Shape (H, W, C):", img.shape)
print("Height:", img.shape[0])
print("Width:", img.shape[1])
print("Channels:", img.shape[2])
print("Data type:", img.dtype)
print("Total pixels:", img.size)


```

```python
# copy image and save it


img_copy = img.copy()
draw = img_copy

# Draw a rectangle and convert into grayscale (copy)
cv2.rectangle(draw, (100,100), (600,600), (0,0,255), 3)
cv2.imwrite("car_copy.png", img_copy)
plt.imshow(cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB))
plt.title("Copied Image (Modified!)")



```

```python
# convert copy into grayscale

grayscale_image = cv2.imread('car_copy.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(grayscale_image)
cv2.imwrite("car_grayscale.png", grayscale_image)
```

```python
_, binary_image = cv2.threshold(grayscale_image, 128, 255, cv2.THRESH_BINARY)
plt.imshow(cv2.cvtColor(binary_image, cv2.COLOR_BGR2RGB))
```
