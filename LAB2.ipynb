```python
import cv2
import matplotlib.pyplot as plt

image_path = "mesi.jpg"

# get image
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# quick operations
images = [
    image,
    cv2.add(image, 100),
    cv2.subtract(image, 100),
    cv2.multiply(image, 2),
    cv2.divide(image, 2)
]

titles = [
    "Original",
    "Brighter (+100)",
    "Darker (-100)",
    "Contrast x2",
    "Divided /2"
]

# plot
plt.figure(figsize=(12, 6))



for i, (img, title) in enumerate(zip(images, titles)):
    plt.subplot(2, 3, i + 1)
    plt.imshow(img)
    plt.title(title)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
